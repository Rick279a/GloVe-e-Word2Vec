** Word2Vec Relacao Entre Palavras **

```python
import gensim.downloader as api
```

```python
# Carregar embeddings Word2Vec pré-treinados
w2v_model = api.load("word2vec-google-news-300")
```

```python
# Obter vetor de palavra para uma palavra específica
word_vector = w2v_model['word']
print("Vetor Word2Vec para a palavra 'word':", word_vector)
```

```python
# Encontrar palavras mais similares
similar_words = w2v_model.most_similar('word', topn=5)
print("Palavras mais similares a 'word':", similar_words)
```
